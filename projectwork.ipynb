Customer Behavior & Demographics

What is the average purchase price, and what are the most common price points?

```python
avg_price=data['Purchase Price'].mean()
avg_price
```

```python
price_count=data['Purchase Price'].value_counts().head(10).reset_index()
price_count.columns=['purchase price','count']
price_count
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=price_count['purchase price'], y=price_count['count'])

plt.xlabel("Purchase Price")
plt.ylabel("Count")
plt.title("Top 10 Most Common Purchase Prices")
plt.xticks(rotation=45)  
plt.show()
```

                                        Summary of the Chart:
The x-axis represents the purchase prices.
The y-axis represents the count (frequency) of each purchase price.
The bars represent how many times each purchase price appears in the dataset.
Some prices (like 49.73 and 89.03) appear more frequently than others.

Key Insights:

The most frequent purchase price is 49.73, appearing 7 times.
Other frequent prices include 5.48, 89.03, and 93.64 (each appearing around 6 times).
The frequency is fairly even, except for a few peak

The box plot shows the distribution of purchase prices for each time period.

Key Insights:

The median purchase price is around 50.33.
The interquartile range (IQR) (middle 50% of values) is between 50.25 and 50.40.
The whiskers extend from approximately 50.19 to 50.50, indicating the range of most values.
There are no significant outliers, meaning most purchase prices are close together
